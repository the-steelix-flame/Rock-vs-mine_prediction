Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data loading and preprocessing


```python
sonar_data = pd.read_csv('/content/Copy of sonar data (1).csv',header=None)
```

```python

sonar_data.describe()
```

```python
sonar_data.head()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()  #grouping the data by the mean of the rocks and the mines separately
```

Now splitting the data and label columns

```python
X = sonar_data.drop(columns=60,axis=1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

Trainng and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=2)
```

```python
print(X,X_train,X_test)
```

Model Training ---> Logistic Regression model

```python
model = LogisticRegression()
```

```python
#Training the logistic regression data with training data
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#accuracy for training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print('accuracy on training data: ',training_data_accuracy)
```

```python
#accuracy for the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print('accuracy on test data: ',test_data_accuracy)
```
